```python
df = pd.read_csv("/content/drive/MyDrive/Notebooks/auto-mpg.csv")
df.head(10)
```

```python
df.drop(['car name'],axis=1,inplace=True)
df.head()
```

```python
df.isnull().sum()
```

```python
df['horsepower'].unique()
```

```python
df.replace('?','0',inplace=True)
```

```python
df['horsepower'].unique()
```

```python
df['horsepower'] = df['horsepower'].astype(int)
```

```python
df.info()
```

```python
df['origin'].unique()
```

```python
x = df.drop('mpg',axis=1)
y = df['mpg']
```

```python
from sklearn.preprocessing import StandardScaler

sc1 = StandardScaler()
y = sc1.fit_transform(np.array(y).reshape(-1,1)).flatten()
y.shape
```

```python
sc2 = StandardScaler()
x = sc2.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
from keras.models import Sequential
from keras.layers import Dense,Dropout,Activation

model = Sequential([
    Dense(128,activation='relu',input_shape=(7,)),
    Dropout(0.2),
    Dense(64,activation='relu'),
    Dropout(0.2),
    Dense(32,activation='relu'),
    Dropout(0.2),
    Dense(1)
])
```

```python
model.compile(optimizer='adam',loss='mse',metrics=['mae','mse'])
```

```python
from keras.callbacks import EarlyStopping,ModelCheckpoint,ReduceLROnPlateau

es = EarlyStopping(monitor='val_loss', patience=20)
lr = ReduceLROnPlateau(monitor='val_loss', factor=0.1, patience=20)
mc = ModelCheckpoint('best_model.keras', monitor='val_loss', save_best_only=True)

model.fit(X_train,y_train,epochs=5000,batch_size=32,callbacks=[es,lr,mc],validation_split=0.2,verbose=1)
```

```python
from keras.models import load_model

model = load_model('best_model.keras')
```

```python
loss,mae,mse = model.evaluate(X_test,y_test)
```
