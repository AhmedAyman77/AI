<h1>Morphological Transformations<h1>
<h4>working with binary images<h4>

```python
img = cv2.imread('smarties.png',0)

# convert image to binary
_, mask = cv2.threshold(img,220,225,cv2.THRESH_BINARY_INV)

kernel = np.ones((5,5),np.uint8)

# image, kernel, iteration
dilation = cv2.dilate(mask,kernel,iterations=2)
erosion = cv2.erode(mask,kernel,iterations=2)
opening = cv2.morphologyEx(mask,cv2.MORPH_OPEN,kernel) # erosion then dilation
closing = cv2.morphologyEx(mask,cv2.MORPH_CLOSE,kernel) # dilation then erosion


cv2.imshow('img', img)
cv2.imshow('mask', mask)
cv2.imshow('dilation', dilation)
cv2.imshow('erosion', erosion)
cv2.imshow('opening', opening)
cv2.imshow('closing', closing)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
image = cv2.imread('smarties.png',0)

# Define a kernel (structuring element)
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (5, 5))

# Top-hat Transform: original image - opening
top_hat = cv2.morphologyEx(image, cv2.MORPH_TOPHAT, kernel)

# Black-hat Transform: closing - original image
black_hat = cv2.morphologyEx(image, cv2.MORPH_BLACKHAT, kernel)

# Morphological Gradient: dilation - erosion
gradient = cv2.morphologyEx(image, cv2.MORPH_GRADIENT, kernel)

# Display results
titles = ['Original Image', 'Top-hat Transform', 'Black-hat Transform', 'Morphological Gradient']
images = [image, top_hat, black_hat, gradient]

plt.figure(figsize=(10, 7))
for i in range(4):
    plt.subplot(2, 2, i + 1)
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.axis('off')

plt.tight_layout()
plt.show()
```
