segmentation method helps us to find out the best threshold value for a given image.

```python
import cv2
import matplotlib.pyplot as plt
import os
os.chdir('./data/')
```

```python
img = cv2.imread('smarties.png',0)

ret,thresh1 = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
ret,thresh2 = cv2.threshold(img,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU) # 0 mean i don't know the threshold value
titles = ['Original Image','BINARY','OTSU']
images = [img, thresh1, thresh2]

for i in range(3):
    plt.subplot(1,3,i+1),plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```
