```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import os
os.chdir('./data/')
```

```python
cap = cv2.VideoCapture(0) # 0 mean record video
# or
# cap = cv2.VideoCapture("path to your video")

fourcc = cv2.VideoWriter_fourcc(*"XVID")

# video name, fourcc, frames (how many photos i want to show per second), (width,height)
out = cv2.VideoWriter('output.avi', fourcc, 20.0, (640, 480))

while cap.isOpened():
    # return ret (bool) , frame if ret = False there is no frame
    ret, frame = cap.read()

    if ret == True:
        print(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
        print(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

        out.write(frame)

        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        cv2.imshow('frame', gray)

        if cv2.waitKey(1) == ord('q'): # is i press q then break
            break
    else:
        break;

cap.release() # close the camera
out.release() # close the video
cv2.destroyAllWindows()
```

<h3>Change Width&Height</h3>

```python
cap = cv2.VideoCapture(0, cv2.CAP_V4L2)  # Use Video4Linux instead of GStreamer

fourcc = cv2.VideoWriter_fourcc(*'XVID')
out = cv2.VideoWriter('output.avi', fourcc, 20.0, (640, 480))

while cap.isOpened():
    ret, frame = cap.read()

    if ret:
        cap.set(3, 3000)  # width
        cap.set(4, 3000)  # height
        print(cap.get(3))
        print(cap.get(4))

        out.write(frame)

        color = cv2.cvtColor(frame, cv2.COLOR_RGB2BGR)
        cv2.imshow('frame', color)
        if cv2.waitKey(1) == ord('q'):
            break

    else:
        break

cap.release()
out.release()
cv2.destroyAllWindows()
```

<h1>put text on video</h1>


```python
import datetime

cap = cv2.VideoCapture(0)

fourcc = cv2.VideoWriter_fourcc(*'XVID')


while cap.isOpened():
    ret, frame = cap.read()
    if ret == True:
        print(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
        print(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
        print(cap.get(3))
        print(cap.get(4))

        font = cv2.FONT_HERSHEY_SIMPLEX
        text = f"Width: {cap.get(3)} Height: {cap.get(4)}"
        date = str(datetime.datetime.now())

        # img, text, (x, y), font, fontScale, color, thickness, lineType
        frame = cv2.putText(frame, date, (10, 50), font, 1, (0, 255, 255), 2, cv2.LINE_AA)
        frame = cv2.putText(frame, text, (10, 100), font, 1, (0, 255, 255), 2, cv2.LINE_AA)

        color = cv2.cvtColor(frame, cv2.COLOR_RGB2BGR)
        cv2.imshow('frame', frame)
        if cv2.waitKey(1) == ord('q'):
            break
    else:
        break

cap.release()
cv2.destroyAllWindows()
```
